```python
#1.0: Retrieve a list of employee_id, first_name, hire_date and department of all employees orderes by the hire date

%sql SELECT employee_id, first_name, department, hire_date, ROW_NUMBER() OVER(ORDER BY hire_date) AS row_n FROM employees;
```

```python
# 1.1: Retrieve the employee_id, first_name, hire_date of employees for different departments ordered by hire date

%sql SELECT employee_id, first_name, hire_date, department, ROW_NUMBER() OVER(PARTITION BY department ORDER BY hire_date) AS row_n FROM employees;
```

## Ranking

```python
# 2.1: Retrieve the employees first_name, email and salary of the highest earner of each department

%sql SELECT first_name, email, department, salary, ROW_NUMBER() OVER(PARTITION BY department ORDER BY salary DESC) FROM employees;
```

```python
# 2.2: Use the RANK function instead

%sql SELECT first_name, email, department, salary, RANK() OVER(PARTITION BY department ORDER BY salary DESC) AS Rank_N FROM employees 
```

```python
# 2.3:  Retrieve the hire date. Return details of employees hired on or before 31st Dec, 2005 and are in first aid, movies and computers departments

%sql SELECT first_name, email, salary, hire_date, department, RANK() OVER(PARTITION BY department ORDER BY salary DESC) FROM employees WHERE hire_date<= '2005-12-31' AND (department='First Aid' OR department='Computers' OR department='Movies');
```
